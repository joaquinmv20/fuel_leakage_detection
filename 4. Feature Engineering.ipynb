```python
# dat50g

columns = ['Unnamed: 0', 'FOB_dif', 'Cluster', 'VALUE_FOB_vs_CONSUMPTION', 'DIFF_VALUE_FOB_vs_FROM_CONSUMPTION']
dat50g = dat50g.drop(columns, axis = True)

dat50g = dat50g.drop(['VALUE_FOB'], axis = True)
dat50g = dat50g.drop(['FLIGHT_PHASE_COUNT'], axis = True)
dat50g['Smoothed_Value'] = dat50g['Simulated_FOB']
dat50g = dat50g.drop(['Simulated_FOB'], axis = True)
dat50g['Flight'] = dat50g['Flight'] + '05'
dat50g = dat50g.drop(['LEAKAGE'], axis = True)
dat50g['FOB_DIF'] = dat50g.groupby('Flight')['Smoothed_Value'].diff()
dat50g['FOB_vs_CONSUMPTION'] = dat50g['Smoothed_Value'] - dat50g['FOB_CONSUMPTION']
dat50g['FOB_vs_CONSUMPTION_DIF'] = dat50g.groupby('Flight')['FOB_vs_CONSUMPTION'].diff()
dat50g.rename(columns={'Smoothed_Value': 'FOB'}, inplace=True)
dat50g['LEAK'] = 1
dat50g.to_csv('msn02_50g.csv', sep = ',')
```
